```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
## Load in csv.gz
info = pd.read_csv('Data/Medical Dataset.csv')
info.head()
info.info()
```

## VitD_levels

### Mean (Red)

```python
col_vid = 'VitD_levels'
mean = info[col_vid].mean()
mean
```

```python
fig,ax = plt.subplots(figsize=(10,6))
sns.histplot(info[col_vid], ax = ax, stat='probability');
ax.axvline(mean, color = 'red')
ax.set_title(col_vid)
ax.legend();
```

### Median (Green)

```python
feature = info[col_vid]
mean = feature.mean()
median = feature.median()
fig,ax = plt.subplots(figsize=(10,6))
sns.histplot(feature ,ax=ax,stat='probability')
ax.axvline(mean, color = 'red',label=f'Mean = {mean:,.2f}')
ax.axvline(median, ls='--', color = 'green', label = f'Median = {median:,.2f}')
ax.set_title(f'{col_vid}')
ax.legend();
```

### +1 std from mean(Black)

```python
feature = info[col_vid]
mean = feature.mean()
median = feature.median()
std = feature.std()
plus_one_std = mean + std
fig,ax = plt.subplots(figsize=(10,6))
sns.histplot(feature ,ax=ax,stat='probability')
ax.axvline(plus_one_std, color = 'black',label=f'+1 std = {plus_one_std:,.2f}')
ax.set_title(f'{col_vid}')
ax.legend();
```

### -1 std from mean(Black)

```python
feature = info[col_vid]
mean = feature.mean()
median = feature.median()
std = feature.std()
minus_one_std = mean - std
fig,ax = plt.subplots(figsize=(10,6))
sns.histplot(feature ,ax=ax,stat='probability')
ax.axvline(minus_one_std, color = 'black', label = f'-1 std = {minus_one_std:,.2f}')
ax.set_title(f'{col_vid}')
ax.legend();
```

### Highlight the range between +1 and =1 std (yellow)

```python
feature = info[col_vid]
mean = feature.mean()
median = feature.median()
std = feature.std()
plus_one_std = mean + std
minus_one_std = mean - std
fig,ax = plt.subplots(figsize=(10,6))
sns.histplot(feature ,ax=ax,stat='probability')
ax.axvline(plus_one_std, color = 'black',label=f'+1 std = {plus_one_std:,.2f}')
ax.axvline(minus_one_std, color = 'black', label = f'-1 std = {minus_one_std:,.2f}')
ax.axvspan(plus_one_std, minus_one_std, color = 'yellow', zorder = 0)
ax.set_title(f'{col_vid}')
ax.legend();
```

## Answer the following questions

### Is it Discrete or Continuous?
This is a Continuous it doesn't have any gaps between the bars in all the histoplots above.

### Does it have a skew? If so, which direction (+/-)
Yes, it has a Negative Skew.

### What type of kurtosis does it display? (Mesokurtic, Leptokurtic, Platykurtic)
The hisplots display a Mesokurtic Kurtosis.
